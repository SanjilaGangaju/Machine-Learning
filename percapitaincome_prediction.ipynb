```python
PREDICTION OF PER CAPITA INCOME OF CANADA IN YEAR 2020
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.head()
```

```python
%matplotlib inline 
plt.scatter(df.year,df['per capita income (US$)'],color='pink',marker='+')
plt.ylabel('capita(US$')
plt.xlabel('year')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline 
plt.scatter(df.year,df['per capita income (US$)'],color='pink',marker='+')
plt.ylabel('capita(US$')
plt.xlabel('year')
plt.plot(df.year,reg.predict(df[['year']]),color='pink')
```
